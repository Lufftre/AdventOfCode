```python
from copy import deepcopy
```

```python
i = """....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""
```

```python
grid = [ [c for c in row] for row in i.splitlines() ]
pos = 0, 0
dir = 0
start_pos = 0, 0
width, height = len(grid[0]), len(grid)
for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if c == "^":
            pos = y, x
            start_pos = pos
            grid[y][x] = "X"

dirs = [
    (-1, 0),
    (0, 1),
    (1, 0),
    (0, -1),
]
while True:
    new_pos = pos[0] + dirs[dir][0], pos[1] + dirs[dir][1]
    if new_pos[0] < 0 or new_pos[0] >= height:
        break
    if new_pos[1] < 0 or new_pos[1] >= width:
        break
    
    tile = grid[new_pos[0]][new_pos[1]]
    if tile == "#":
        dir = (dir + 1) % 4
        continue
    grid[new_pos[0]][new_pos[1]] = "X"
    pos = new_pos

s = 0
for r in grid:
    for c in r:
        if c == "X":
            s += 1
print(s)
grid2 = grid.copy()
```

```python
def solve(grid):
    pos = start_pos
    dir = 0
    history = set()
    while True:
        new_pos = pos[0] + dirs[dir][0], pos[1] + dirs[dir][1]
        if (new_pos, dir) in history:
            return True
        history.add((new_pos, dir))
        
        if new_pos[0] < 0 or new_pos[0] >= height:
            return False
        if new_pos[1] < 0 or new_pos[1] >= width:
            return False
        tile = grid[new_pos[0]][new_pos[1]]
        if tile == "#":
            dir = (dir + 1) % 4
            continue
        pos = new_pos
```

```python
s2 = 0
for y, r in enumerate(grid2):
    for x, c in enumerate(r):
        if c == "X" and (y, x) != start_pos:
            tmp_grid = deepcopy(grid2)
            tmp_grid[y][x] = "#"
            s2 += solve(tmp_grid)
print(s2)
```
