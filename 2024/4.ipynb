```python
def serach_right(row, x):
    try:
        if row[x + 1] != "M": return False
        if row[x + 2] != "A": return False
        if row[x + 3] != "S": return False
        return True
    except:
        return False
def serach_left(row, x):
    if x < 3: return False
    try:
        if row[x - 1] != "M": return False
        if row[x - 2] != "A": return False
        if row[x - 3] != "S": return False
        return True
    except:
        return False
def serach_up(y, x):
    if y < 3: return False
    try:
        if grid[y - 1][x] != "M": return False
        if grid[y - 2][x] != "A": return False
        if grid[y - 3][x] != "S": return False
        return True
    except:
        return False
def serach_down(y, x):
    try:
        if grid[y + 1][x] != "M": return False
        if grid[y + 2][x] != "A": return False
        if grid[y + 3][x] != "S": return False
        return True
    except:
        return False
def serach_diag_upright(y, x):
    if y < 3: return False
    try:
        if grid[y - 1][x + 1] != "M": return False
        if grid[y - 2][x + 2] != "A": return False
        if grid[y - 3][x + 3] != "S": return False
        return True
    except:
        return False
def serach_diag_upleft(y, x):
    if x < 3: return False
    if y < 3: return False
    try:
        if grid[y - 1][x - 1] != "M": return False
        if grid[y - 2][x - 2] != "A": return False
        if grid[y - 3][x - 3] != "S": return False
        return True
    except:
        return False
def serach_diag_downright(y, x):
    try:
        if grid[y + 1][x + 1] != "M": return False
        if grid[y + 2][x + 2] != "A": return False
        if grid[y + 3][x + 3] != "S": return False
        return True
    except:
        return False
def serach_diag_downleft(y, x):
    if x < 3: return False
    try:
        if grid[y + 1][x - 1] != "M": return False
        if grid[y + 2][x - 2] != "A": return False
        if grid[y + 3][x - 3] != "S": return False
        return True
    except:
        return False
found = 0
grid = i.splitlines()
for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if c == "X":
            found += serach_left(row, x)
            found += serach_right(row, x)
            found += serach_up(y, x)
            found += serach_down(y, x)
            found += serach_diag_downleft(y, x)
            found += serach_diag_downright(y, x)
            found += serach_diag_upleft(y, x)
            found += serach_diag_upright(y, x)
print(found)
```

```python
def serach_diag_upright(y, x):
    y += 1
    x -= 1
    if y < 3: return False
    try:
        if grid[y - 1][x + 1] != "M": return False
        if grid[y - 2][x + 2] != "A": return False
        if grid[y - 3][x + 3] != "S": return False
        return True
    except:
        return False
def serach_diag_upleft(y, x):
    y += 1
    x += 1
    if x < 3: return False
    if y < 3: return False
    try:
        if grid[y - 1][x - 1] != "M": return False
        if grid[y - 2][x - 2] != "A": return False
        if grid[y - 3][x - 3] != "S": return False
        return True
    except:
        return False
def serach_diag_downright(y, x):
    y -= 1
    x -= 1
    try:
        if grid[y + 1][x + 1] != "M": return False
        if grid[y + 2][x + 2] != "A": return False
        if grid[y + 3][x + 3] != "S": return False
        return True
    except:
        return False
def serach_diag_downleft(y, x):
    y -= 1
    x += 1
    if x < 3: return False
    try:
        if grid[y + 1][x - 1] != "M": return False
        if grid[y + 2][x - 2] != "A": return False
        if grid[y + 3][x - 3] != "S": return False
        return True
    except:
        return False
found = 0
grid = i.splitlines()
for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if c == "A":
            a = serach_diag_downleft(y - 1, x + 1) or serach_diag_upright(y + 1, x - 1)
            b = serach_diag_downright(y - 1, x - 1) or serach_diag_upleft(y + 1, x + 1)
            found += a and b

print(found)
```
