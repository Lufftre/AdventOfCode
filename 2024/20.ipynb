```python
def init():
    grid = []
    for y, row in enumerate(input.splitlines()):
        r = []
        for x, c in enumerate(row):
            if c == "S":
                start = (x,y)
            if c == "E":
                end = (x,y)
            r.append(c)
        grid.append(r)
    return grid, start, end
grid, start, end = init()
```

```python
def get_neighbours(pos):
    ns = []
    for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        xx = pos[0] + dx
        yy = pos[1] + dy
        ns.append((xx, yy))
    return ns

def get(pos):
    return grid[pos[1]][pos[0]]
```

```python
seen = {}
walls = {}
def search(pos, score):
    frontier = [(pos, score)]
    while len(frontier):
        pos,score = frontier.pop()
        seen[pos] = score
        ns = get_neighbours(pos)
        for n in ns:
            c = get(n)
            if c == "#":
                walls[n] = walls.get(n, 0) + 1
            elif n not in seen:
                frontier.append((n, score + 1))

search(start, 0)
print(seen[end])
```

```python
shortcuts = []
for wall, s in walls.items():
    if s == 2:
        shortcuts.append(wall)

deltas = []
for s in shortcuts:
    ns = get_neighbours(s)
    scores = []
    for n in ns:
        ss = seen.get(n)
        if ss is not None:
            scores.append(ss)
    if len(scores) != 2:
        print(scores)
    ss = abs(scores[0] - scores[1])
    deltas.append(ss - 2)

s = 0
for d in deltas:
    if d >= 100:
        s += 1
print(s)
```

```python
def manhattan(pos, pos2):
    dx = pos[0] - pos2[0]
    dy = pos[1] - pos2[1]
    return abs(dx) + abs(dy)
```

```python
deltas = []
for pos, s in seen.items():
    for pos2, s2 in seen.items():
        m = manhattan(pos, pos2)
        if  m > 1 and m <= 20:
            deltas.append(s2 - s - m)

s = 0
for d in deltas:
    if d >= 100:
        s += 1
print(s)
```
