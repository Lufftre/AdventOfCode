```python
grid = i.splitlines()
w, h = len(grid[0]), len(grid)
def dfs(x: int, y: int, seen: set):
    v = grid[y][x]
    if v == "9":
        return 1
    
    s = 0
    for xx, yy in [(x+1, y), (x-1, y), (x, y+1), (x, y-1)]:
        if (xx, yy) in seen:
            continue
        if xx < 0 or xx >= w:
            continue
        if yy < 0 or yy >= h:
            continue
        if int(grid[yy][xx]) != int(v) + 1:
            continue
        seen2 = seen.copy()
        seen2.add((xx, yy))
        s += dfs(xx, yy, seen2)
    return s

s = 0
for y,r in enumerate(grid):
    for x,c in enumerate(r):
        if c == "0":
            s += dfs(x, y, set())
print(s)
```
