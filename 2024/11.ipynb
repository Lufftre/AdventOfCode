```python
from collections import Counter

stones = Counter(i.split())
def change_stone(stone: str):
    if stone == "0":
        return ["1"]
    
    l = len(stone)
    if l % 2 == 0:
        return [stone[:l//2], str(int(stone[l//2:]))]
    
    return [str(int(stone) * 2024)]

def step(stones: dict):
    new_stones = {}
    for key, val in stones.items():
        new = change_stone(key)
        for s in new:
            new_stones[s] = new_stones.get(s, 0) + val
    return new_stones

for x in range(75):
    stones = step(stones)

s = 0
for n in stones.values():
    s += n
print(s)
```
