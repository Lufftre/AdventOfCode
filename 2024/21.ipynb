```python
class Numpad:
    grid = [" 0A", "123", "456", "789"]

    def __init__(self):
        self.pos = (2, 0)
        self.button = "A"
    
    @staticmethod
    def keyi_to_pos(keyi):
        keyi -= 1

        x = keyi % 3
        y = keyi // 3 + 1
        return (x, y)

    @staticmethod
    def key_to_keyi(key):
        if key == "A":
            keyi = 0
        elif key == "0":
            keyi = -1
        else:
            keyi = int(key)
        return keyi
    
    @staticmethod
    def key_to_pos(key):
        keyi = Numpad.key_to_keyi(key)
        pos = Numpad.keyi_to_pos(keyi)
        return pos

    @staticmethod
    def pos_to_key(pos):
        x, y = pos
        return Numpad.grid[y][x]   

    def path(self, pos1, pos2):
        x, y = pos1
        xx, yy = pos2
        dx = xx - x
        dy = yy - y
        path = ""
        tokenx = ">" if dx > 0 else "<"
        tokeny = "^" if dy > 0 else "v"
        if dx > 0:
                path += tokenx*abs(dx)
                path += tokeny*abs(dy)
        else:
                path += tokeny*abs(dy)
                path += tokenx*abs(dx)
        self.pos = pos2
        self.button = self.pos_to_key(pos2)
        return path
    
    
```

```python
#     +---+---+
#     | ^ | A |
# +---+---+---+
# | < | v | > |
# +---+---+---+
class Keypad:
    grid = ["<v>", " ^A"]
    def __init__(self):
        self.pos = (2, 1)
        self.button = "A"
    
    @staticmethod
    def keyi_to_pos(keyi):
        x = keyi % 3
        y = keyi // 3 
        return (x, y)

    @staticmethod
    def key_to_keyi(key):
        if key == "A":
            return 5
        if key == "^":
             return 4
        if key == ">":
             return 2
        if key == "v":
             return 1
        if key == "<":
             return 0

    @staticmethod
    def key_to_pos(key):
        keyi = Keypad.key_to_keyi(key)
        pos = Keypad.keyi_to_pos(keyi)
        return pos
    
    @staticmethod
    def pos_to_key(pos):
        x, y = pos
        return Keypad.grid[y][x]    
    
    def path(self, pos1, pos2):
        x, y = pos1
        xx, yy = pos2
        dx = xx - x
        dy = yy - y
        path = ""
        tokenx = ">" if dx > 0 else "<"
        tokeny = "^" if dy > 0 else "v"
        if dx < 0:
                path += tokeny*abs(dy)
                path += tokenx*abs(dx)
        else:
                path += tokenx*abs(dx)
                path += tokeny*abs(dy)
        self.pos = pos2
        self.button = self.pos_to_key(pos2)
        return path


```

```python
import re
def complexity(n, code):
    c = int(re.sub('[^0-9]', '', code))
    print(n, c)
    return n * c

numpad = Numpad()
keypad1 = Keypad()
keypad2 = Keypad()

codes = [
    "413A",
    "480A",
    "682A",
    "879A",
    "083A",
]

codes = [
    "029A",
    "980A",
    "179A",
    "456A",
    "379A",
]
tot = 0
for code in codes:
    robot1 = ""
    robot2 = ""
    my_path = ""
    for c in code:
        pos1 = numpad.key_to_pos(c)
        path1 = numpad.path(numpad.pos, pos1) + "A"
        robot1 += path1

    for c in robot1:
        pos1 = keypad1.key_to_pos(c)
        path1 = keypad1.path(keypad1.pos, pos1) + "A"
        robot2 += path1

    for c in robot2:
        pos1 = keypad2.key_to_pos(c)
        path1 = keypad2.path(keypad2.pos, pos1) + "A"
        my_path += path1

    comp = complexity(len(my_path), code)
    print(code, comp)
    print()
    tot += comp
print(tot)
```
