```python
# +---+---+---+
# | 7 | 8 | 9 |
# +---+---+---+
# | 4 | 5 | 6 |
# +---+---+---+
# | 1 | 2 | 3 |
# +---+---+---+
#     | 0 | A |
#     +---+---+
# ---------------

class Numpad:
    nogo = (0,0)
    grid = [" 0A", "123", "456", "789"]

    def __init__(self):
        self.pos = (2, 0)
        self.button = "A"
    
    @staticmethod
    def keyi_to_pos(keyi):
        keyi -= 1

        x = keyi % 3
        y = keyi // 3 + 1
        return (x, y)

    @staticmethod
    def key_to_keyi(key):
        if key == "A":
            keyi = 0
        elif key == "0":
            keyi = -1
        else:
            keyi = int(key)
        return keyi
    
    @staticmethod
    def key_to_pos(key):
        keyi = Numpad.key_to_keyi(key)
        pos = Numpad.keyi_to_pos(keyi)
        return pos

    @staticmethod
    def pos_to_key(pos):
        x, y = pos
        return Numpad.grid[y][x]   

    def path1(self, pos1, pos2):
        x, y = pos1
        xx, yy = pos2
        dx = xx - x
        dy = yy - y
        vx = 1 if dx > 0 else -1
        vy = 1 if dy > 0 else -1
        path = []
        tokenx = ">" if dx > 0 else "<"
        tokeny = "^" if dy > 0 else "v"
        for i in range(abs(dx)):
             x += vx
             if (x,y) == Numpad.nogo:
                  return None
             path.append(tokenx)
        for i in range(abs(dy)):
             y += vy
             if (x,y) == Numpad.nogo:
                  return None
             path.append(tokeny)
        return ''.join(path)

    
    def path2(self, pos1, pos2):
        x, y = pos1
        xx, yy = pos2
        dx = xx - x
        dy = yy - y
        vx = 1 if dx > 0 else -1
        vy = 1 if dy > 0 else -1
        path = []
        tokenx = ">" if dx > 0 else "<"
        tokeny = "^" if dy > 0 else "v"
        for i in range(abs(dy)):
            y += vy
            if (x,y) == Numpad.nogo:
                return None
            path.append(tokeny)
        for i in range(abs(dx)):
            x += vx
            if (x,y) == Numpad.nogo:
                return None
            path.append(tokenx)
        return ''.join(path)


    def all_paths(self, pos1, pos2):
        valid = []
        p1 = self.path1(pos1, pos2)
        p2 = self.path2(pos1, pos2)
        if p1 is not None:
             valid.append(p1)
        if p2 is not None and p1 != p2:
             valid.append(p2)
        return valid

    def path(self, pos1, pos2):
        x, y = pos1
        xx, yy = pos2
        dx = xx - x
        dy = yy - y
        path = ""
        tokenx = ">" if dx > 0 else "<"
        tokeny = "^" if dy > 0 else "v"
        if dx > 0:
                path += tokenx*abs(dx)
                path += tokeny*abs(dy)
        else:
                path += tokeny*abs(dy)
                path += tokenx*abs(dx)
        self.pos = pos2
        self.button = self.pos_to_key(pos2)
        return path
    
    
```
